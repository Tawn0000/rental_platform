```python
def get_statics(key, option=0):
    if key == 'address':
        key_set = set()
        for item_list in df.take(total_num):
            key_set.add(item_list[key][option])
    elif key == 'facility':
        key_set = set()
        for item_list in df.take(total_num):
            if not item_list[key]:
                continue
            for item in item_list[key]:
                key_set.add(item)
    elif key == 'tag':
        key_set = {item[:5] for item_list in df.take(total_num) for item in item_list[key]}
    elif key == 'house_type':
        key_set = {item[key].replace(' ','') for item in df.take(total_num)}
    else:
        key_set = {item[key] for item in df.take(total_num)}
    print(key_set)
    if key == 'price':
        key_list = [int(float(item)) for item in list(key_set)]
        key_list.sort()
    else:
        key_list = list(key_set)
    
    key_dict = {}
    for item in key_list:
        key_dict[item] = 0

    if key == 'price':
        for item in df.take(total_num):
            key_dict[int(float(item[key]))] += 1
    elif key == 'address':
        for item_list in df.take(total_num):
                key_dict[item_list[key][option]] += 1
    elif key == 'facility':
        for item_list in df.take(total_num):
            if not item_list[key]:
                continue
            for item in item_list[key]:
                key_dict[item] += 1
    elif key == 'tag':
        for item_list in df.take(total_num):
            for item in item_list[key]:
                key_dict[item[:5]] += 1
    elif key == 'house_type':
        for item in df.take(total_num):
            key_dict[item[key].replace(' ','')] += 1
    else:
        for item in df.take(total_num):
            key_dict[item[key]] += 1
    s = 0
    for (k,v) in key_dict.items():
        print(k, ':', v)
        s += v
    print('totall: ',s)
```

## 调用高德地图API求经纬度坐标和步行距离

```python
import requests
def get_coordination(keywords,city='无锡'):
    url = 'https://restapi.amap.com/v3/assistant/inputtips?output=json&city=%s&keywords=%s&key=7d584a843bb546ac822c4c668ff25b8f'
    url = url % (city, keywords)
    response = requests.get(url)
    answer = response.json()
    print(url)
#     print(answer)
    if answer.get('status') == '1' and answer.get('count') != '0':
        print(answer.get('tips')[0].get('name'))
        return answer.get('tips')[0].get('location')
    
def get_distance(origin, destination):
    print(origin)
    print(destination)
    url = 'https://restapi.amap.com/v3/direction/walking?origin=%s&destination=%s&key=7d584a843bb546ac822c4c668ff25b8f'
    url = url % (origin, destination)
    response = requests.get(url)
    answer = response.json()
    print(url)
    print(answer)
    
address1 = '无锡江南大学李园'
address2 = '无锡江南大学物联网工程学院'
get_distance(get_coordination(address1), get_coordination(address2))
```
